```
answer7 = int( round( (df.successful_orders.sum()) / df.vendor_id.nunique() ) )
print(answer7)
```

```
answer8 = int(df.groupby('city_id').successful_orders.sum().reset_index().nsmallest(1, 'successful_orders').iloc[0].city_id)
print(answer8)
```

```
successful_orders = df.groupby('spec').successful_orders.sum().reset_index()
fail_orders = df.groupby('spec').fail_orders.sum().reset_index()
df_orders = pd.merge(successful_orders, fail_orders, on='spec')
df_orders['part_fail'] = df_orders['fail_orders'] / (df_orders['successful_orders']+ df_orders['fail_orders'])
answer9 = df_orders.nlargest(1, 'part_fail').iloc[0].spec
print(answer9)
```

```
was_in_july = df[(df.date < '2019-08-01') & (df.date > '2019-07-00')].groupby('chain_id').successful_orders.sum().reset_index()
was_in_aug = df[(df.date < '2019-09-01') & (df.date > '2019-08-00')].groupby('chain_id').successful_orders.sum().reset_index()
df_july_aug = pd.merge(was_in_july, was_in_aug, on='chain_id', how='inner')
df_july_aug['abs'] = np.abs(df_july_aug.successful_orders_x - df_july_aug.successful_orders_y)
answer10 = int(df_july_aug.nlargest(1, 'abs').iloc[0].chain_id)
print(answer10)
```
